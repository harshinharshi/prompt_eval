# Test Groq Integration with LangChain

This notebook tests the integration of Groq with LangChain to ensure everything is working correctly.

## Section 1: Import Required Libraries

```python
import os
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()
```

## Section 2: Initialize Groq API Client

```python
# Get the API key from environment
api_key = os.getenv("GROQ_API_KEY")

# Check if API key is available
if not api_key:
    print("ERROR: GROQ_API_KEY not found in environment variables!")
    print("Please make sure your .env file has GROQ_API_KEY set.")
else:
    print("✓ GROQ_API_KEY found in environment")

# Initialize the Groq ChatGroq client
llm = ChatGroq(
    groq_api_key=api_key,
    model="llama-3.3-70b-versatile",
    temperature=0.7,
)

print("✓ Groq LLM client initialized successfully")
```

## Section 3: Create and Send Test Message

```python
# Create a simple message
message = "how are you"

print(f"\n{'='*60}")
print(f"SENDING MESSAGE TO GROQ:")
print(f"{'='*60}")
print(f"Message: {message}")
print(f"{'='*60}\n")

# Create a HumanMessage and invoke the LLM
messages = [HumanMessage(content=message)]

try:
    # Get response from Groq
    response = llm.invoke(messages)
    
    print(f"{'='*60}")
    print(f"RESPONSE FROM GROQ:")
    print(f"{'='*60}")
    print(response.content)
    print(f"{'='*60}\n")
    
    print("✓ SUCCESS! Groq integration is working correctly!")
    
except Exception as e:
    print(f"❌ ERROR: {str(e)}")
    print("Failed to get response from Groq")
```
